#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5 net revenue generated for over all orders place is 7601283.25

```python
def category_price(dp):
    if dp > 600:
        return (dp - (dp * 0.25))
    elif 350 < dp < 600:
        return (dp - (dp * 0.15))
    elif 100 < dp < 350:
        return (dp - (dp * 0.10))
    else:
        return (dp - (dp * 0.05))

ECom_Data['net_revenue_company'] = ECom_Data['discounted_price'].apply(category_price)
```

```python
ECom_Data
```

```python
ECom_Data['net_revenue_company'].sum()
```
